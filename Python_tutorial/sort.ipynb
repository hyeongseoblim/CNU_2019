```python
def selection_sort(my_list):
    for index in range(len(my_list)-1):
        min_value = index
        for i in range(index+1,len(my_list)):
            if my_list[i] < my_list[min_value]:
                min_value = i
                
        temp = my_list[min_value]
        my_list[min_value] = my_list[index]
        my_list[index] = temp

    return my_list

some_list = [11, 3, 6, 4, 12, 1, 2]
selection_sort(some_list)
print(some_list)

def bubble_sort(my_list):
    for i in range(len(my_list)):
        for j in range(i+1,len(my_list)):
            if my_list[j] < my_list[i]:
                temp = my_list[j]
                my_list[j] = my_list[i]
                my_list[i] = temp
some_list = [11, 3, 6, 4, 12, 1, 2]
insertion_sort(some_list)
print(some_list)

def insertion_sort(my_list):
    for i in range(len(my_list)):
        key = my_list[i]
        for j in range(i+1,len(my_list)):
            if my_list[key] < my_list[j]:
                temp = my_list[key]
                my_list[key] = my_list[j]
                my_list[j]=temp
                
some_list = [11, 3, 6, 4, 12, 1, 2]
insertion_sort(some_list)
print(some_list)

# 삽입 정렬
def insertion_sort(my_list):
    for i in range(len(my_list)):
        key = my_list[i]
        j = i - 1
        while j >= 0 and my_list[j] > key:
            my_list[j + 1] = my_list[j]
            j = j - 1

        my_list[j + 1] = key

some_list = [11, 3, 6, 4, 12, 1, 2]
insertion_sort(some_list)
print(some_list)

def merge(list1, list2):
    new_list = list()
    i=0
    j = 0
    
    while i<len(list1) and j <len(list2):
        if list1[i]<=list2[j]:
            new_list.append(list1[i])
            i += 1
        else:
            new_list.append(list2[j])
            j += 1
        
    for k in list1:
        if k not in new_list:
            new_list.append(k)
    for l in list2:
        if l not in new_list:
            new_list.append(l)
    return new_list
    
# 합병 정렬
def merge_sort(my_list):
   
    if len(my_list) < 2:
        return my_list
    
    length = len(my_list)//2
    f_list = my_list[:length]
    f_list = merge_sort(f_list)
    s_list = my_list[length:]
    s_list = merge_sort(s_list)
    my_list = merge(f_list,s_list)
    
    return my_list
    

some_list = [11, 3, 6, 4, 12, 1, 2]
sorted_list = merge_sort(some_list)
print(sorted_list)
```
