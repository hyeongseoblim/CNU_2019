```python
group_season = train.groupby(['SEASON'])['RENT_COUNT'].sum().reset_index()
ax = sns.barplot(x=group_season['SEASON'], y=group_season['RENT_COUNT'])
ax.set(xlabel='Season',ylabel='Rent count')
season = ['Spring','Summer','Fall','Winter']
plt.xticks(np.arange(4),season)
plt.title('The number of rented bike by season')
plt.show()
```

```python
group_weekday = train.groupby(['WEEKDAY'])['RENT_COUNT'].sum().reset_index()
ax = sns.barplot(x=group_weekday['WEEKDAY'],
                y=group_weekday['RENT_COUNT'])
ax.set(xlabel='Weekday', ylabel='count')
weekday=['SUN','MON','TUE','WED','THU','FRI','SAT']
plt.xticks(np.arange(7),weekday)
plt.title('The number of rented bike by weekday')
plt.show()
```

```python
group_mn = train.groupby(['MONTH'])['RENT_COUNT'].sum().reset_index()
ax = sns.barplot(x=group_mn['MONTH'], y=group_mn['RENT_COUNT'])
ax.set(xlabel='Month',ylabel='count')
plt.title('The number of rented bike by weekday')
plt.show()
```

```python
group_hr = train.groupby(['HOUR'])['RENT_COUNT'].sum().reset_index()
ax = sns.barplot(x=group_hr['HOUR'], y=group_hr['RENT_COUNT'])
ax.set(xlabel='Hour',ylabel='count')
plt.title('The number of rented bike by weekday')
plt.show()
```

```python
plt.figure(figsize=(10,6))
station_3 = train[train['STATION']==3]
group_weekday_hour = station_3.groupby(['WEEKDAY','HOUR'])['RENT_COUNT'].mean().reset_index()
weekday_hour = np.zeros([7,24])
for i in range(7*24):
    weekday_hour[int(group_weekday_hour.iloc[i][0])][int(group_weekday_hour.iloc[i][1])]=group_weekday_hour.iloc[i][2]
ax = sns.heatmap(weekday_hour, cmap="YlGnBu")
ax.set(xlabel='Bour',ylabel='Weekday')
weekday = ['SUN','MON','TUE','WED','THU','FRI','SAT']
plt.yticks(np.arange(7),weekday)
plt.show()
```

```python
import gmplot
import statistics

tas=pd.read_csv('./data/tas.csv')
del tas['RENT_DATE']
del tas['RETURN_DATE']
tas=tas[tas.RETURN_STATION<=144]
tas=tas[tas.RENT_STATION<=144]

tas['COUNT']=0
tas = tas.groupby(['RENT_STATION','RETURN_STATION']).count().reset_index()

top_10 = tas.nlargest(10, 'COUNT')
top_10
```

```python
top_10df
top10_lat=[36.369855,36.361773,36.361736,36.373457,36.374325,36.359293,36.364177,36.352190,36.358494,36.361736,36.369855,36.361773,36.361736,36.364177,36.374325,36.359293,36.373457,36.352190,36.358494,36.359293]
top10_lon=[127.388749,127.357485,127.344994,127.359293,127.387462,127.354503,127.358845,127.378814,127.361197,127.344994,127.388749,127.357485,127.344994,127.358845,127.387462,127.354503,127.359293,127.378814,127.361197,127.354503]
print(len(top10_lat),len(top10_lon))
```

```python
gmapl = gmplot.GoogleMapPlotter(statistics.median(top10_lat),statistics.median(top10_lon),14)

for i in range(10):
    lat=[]
    lon=[]
    lat.append(top10_lat[i])
    lat.append(top10_lat[i+10])
    lon.append(top10_lon[i])
    lon.append(top10_lon[i+10])
    gmapl.scatter(lat, lon, 'blue', size=30,marker=False)
    gmapl.plot(lat,lon,'red',edge_width=2.5)

gmapl.draw('top10.html')
```

```python
top_100 =tas.nlargest(100,'COUNT')

top_100 = top_100.rename(columns={"RENT_STATION":"source","RETURN_STATION":"target","COUNT":"value"})

hv.Chord(top_100)
```

```python
del station_df['키오스크번호']
del station_df['주소']
del station_df['좌표']
del station_df['위치']
del station_df['구별']
station_df = station_df.rename(columns={"번호":"index","거치대":"group","명칭":"name"})

station_df_col =station_df.columns.tolist()
station_df_col=station_df_col+[station_df_col.pop(1)]
station_df = station_df[station_df_col]
station_df=hv.Dataset(station_df,'index',['group','name'])

chord = hv.Chord((top_100, station_df)).select(value=(5, None))
chord.opts(
    opts.Chord(cmap='Category20', edge_cmap='Category20', edge_color=dim('source').str(), 
               labels='name', node_color=dim('index').str()))
```
